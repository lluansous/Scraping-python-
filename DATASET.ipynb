```python
#Dataset:
#SCRAPING USANDO DADOS EM DOCUMENTOS DISPONIBILIZADOS PELO SISTEMA DE APOIO ÀS LEIS DE INCENTIVO À CULTURA EM: "http://sistemas.cultura.gov.br/salicnet/Salicnet/Salicnet.php#"
# - Contendo dados de pontos de cultura(Registos, UF e cidade)
# - Comparativo dos projetos apresentados, apoiados e aprovados pelo  Fundo Nacional Da Cultura e Macenato.(Ano, Quantidade e Valores)
# - Icentivadores (Ano, tipo de pessoa e região)
# - Maiores preponentes
```

```python
#Importanto as bibliotecas para fazer os Scraping 
#Tabula é capaz de  ler tabelas de um PDF e convertê-las em um DataFrame do pandas. 
import  tabula
import pandas as pd
```

```python
#Scraping: comparativo dos projetos
pdf_path3  =  "Comparativo por Ano.pdf"
dfs  =  tabula.read_pdf ( 
    pdf_path3 , 
    pages = "all" , 
    lattice = True , 
    pandas_options = { "header" :  [ 0 ]}, 
    area = [ 0 ,  0 ,  100 ,  100 ], 
    relative_area = True , 
    multiple_tables = False , 
) 
dfs [ 0 ]
```

```python
df = pd.DataFrame(dfs[0])
```

```python
df
```

```python
#Apartir daqui o objetivo é limpar os dados o máximo que puder ou que eu souber haha. 
#Objetivo maior será retirar o cabeçalho, colocar os valores NaN = 0 e renomear as colunas que o scraping conseguiu.
#Salvar em csv. 
```

```python
df.iloc[2]
df.loc[20]
df.info()
```

```python
new_header = df.iloc[0] # pega a primeira linha para o cabeçalho
df = df[1:] #captura os dados menos a linha do cabecalho
df.columns = new_header #definir a linha do cabeçalho como o cabeçalho df
```

```python
df
```

```python
df = df.fillna(0)
```

```python
df.drop(27, axis = 0, inplace = True)
```

```python
df.drop(32, axis = 0, inplace = True)
```

```python
df.rename(columns={'Total\rGeral': 'Total Geral'}, inplace = True)
```

```python
df['Ano'].replace('Total\rGeral', 'Total Geral', inplace = True)
```

```python
df = df.rename({'Qtde\rApres.': 'Quantidade Apresentada'}, axis = 1)
df = df.rename({'Qtde\rAprov.': 'Quantidade Aprovada'}, axis = 1)
df = df.rename({'Qtde\rApoio.': 'Quantidade Apoiada'}, axis = 1)
```

```python
df
```

```python
df.to_csv('Comparativodosprojetos.csv', sep='\t', encoding='utf-8')
```
